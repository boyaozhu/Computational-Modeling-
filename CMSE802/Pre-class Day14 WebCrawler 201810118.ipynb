** For plotting commands, you can investigate draw_networkx series of commands for drawing nodes and edges, and adding labels. **

```python
# Your code here.

# load libraries
import networkx as nx
import matplotlib.pyplot as plt


# Create a directed graph object
FstV = nx.DiGraph()  # <== Fill the blank. There are four types of graph objects: Graph, DiGraph, MultiGraph, MultiDiGraph.
                     # Investigate which one you need for this task. 


# the rest of your code
# add nodes
FstV.add_node(1,pos = (0,0))
FstV.add_node(2,pos = (2,0))
FstV.add_node(3,pos = (2,2))
FstV.add_node(4,pos = (0,2))
FstV.add_node(5,pos = (1,1))



# add edges
FstV.add_edge(1,5, weight=1)
FstV.add_edge(2,5, weight=1)
FstV.add_edge(5,3, weight=3.5)
FstV.add_edge(4,5, weight=0.5)
FstV.add_edge(4,3, weight=0.5)

# get nodes' positions for plotting
pos=nx.get_node_attributes(FstV,'pos')
label = {}
label[1] = "$B$"
label[2] = "$E$"
label[3] = "$D$"
label[4] = "$A$"
label[5] = "$C$"


# plotting
nx.draw(FstV, pos)



labels = nx.get_edge_attributes(FstV,'weight')
nx.draw_networkx_edge_labels(FstV,pos,edge_labels=labels)
nx.draw_networkx_labels(FstV,pos,label,font_size=16)

plt.show()
# calculate pagerank scores

pr = nx.pagerank(FstV,alpha=0.85)
print (pr)
```
