### Part 1: High order upwind scheme
As we have seen in the pre-class assignemnt, the first-order upwind scheme introduces severe numerical diffusion. One way to alleviate such an artifact is to use higher order upwind schemes. In this practice, we will implement the **second** and **third-order** upwind schemes for the advection term. 

### Second-order upwind scheme:
For a positive advenction velocity, the derivative is calculated using three points in backward dfference:

$$u \frac{\partial C}{\partial x} = u_{i} \frac{3C_{i} - 4C_{i-1} + C_{i-2}}{2 \Delta x}.$$

See the figure below for illustration. For a negative advection velocity, the derivative is calculated using three points in forward dfference:

$$u \frac{\partial C}{\partial x} = u_{i} \frac{-C_{i+2} + 4C_{i+1} -3 C_{i}}{2 \Delta x}.$$

In short, the derivative is taken on the side where the wind blows from.

### Now, let's modify our code in the pre-class assignment to include the 2nd-order upwind scheme. During 0 < t < 1.0, the wind blows from the left (u_x = +1). After t > 1.0, the wind blows from the right (u_x = -1). Complete the code below and do the simulation.

**Since the 2nd-order upwind scheme needs to use two neighboring grid points from the point where derivative is taken, we will need two layers of ghost grids (grids for boundary conditions). We will only track the evolution of concentration at points excluding those ghost grids. Think about the tally of grid points of interest and ghost grids**

```python
############################################################
# Code 1: 2nd-order upwind scheme
############################################################

import math
import numpy as np
import time
import matplotlib.pyplot as plt
from IPython.display import display, clear_output


# parameters
tm = 0.0
dfs = 1.0e-5
X_0 = 0.8

# spatial discretization
nx = 301

Xg = np.linspace(0,3,nx)
dx = Xg[2]-Xg[1]

dt = 5.0*1e-4

print(dx,dt)

# lists for storing results
Con_o = np.zeros(nx)
Con_n = np.zeros(nx)

Cnvc = np.zeros(nx)
Con_test = np.zeros(nx)

# initial concentration profile 
Con_o[0:nx] = np.exp(-(Xg[0:nx]-X_0)**2/(4*500*dfs))
Con_test[0:nx] = np.exp(-(Xg[0:nx]-X_0)**2/(4*500*dfs))
plt.plot(Xg,Con_test,'-')

# time iteration
for iter in range(4000):
    
    if tm < 1.0:
        # advection velocity
        u_x = +1.0
        # convection part 
        Cnvc[2:nx-2] = u_x*(3*Con_o[2:nx-2]-4*Con_o[1:nx-3]+Con_o[0:nx-4])/(2*dx)    # <== fill the blank
        
    else:
        # advection velocity
        u_x = -1.0
        Cnvc[2:nx-2] = u_x*(-Con_o[4:nx]+4*Con_o[3:nx-1]-3*Con_o[2:nx-2])/(2*dx)    # <== fill the blank
    
    # update concentration
    Con_n[2:nx-2] = Con_o[2:nx-2] - dt*Cnvc[2:nx-2]     # <== fill the blank
    
    # boundary condition
    Con_n[0] = Con_n[1]
    Con_n[1] = Con_n[2]                    # <== fill the blank
    
    tm = tm + dt
    
    Con_o[:] = Con_n[:]
    
    
    # visualization
    if iter%50 == 0:        
        clear_output(wait=True) # Clear output for dynamic display
        
        plt.plot(Xg,Con_o,'-')
        plt.axis([0.0, 3.0, -0.2, 1.2])
        plt.show()
     
        time.sleep(0.01)         # Sleep for half a second to slow down the animation
        
        
print(tm,max(Con_o[2:nx-2]))  

plt.plot(Xg,Con_o,'r.-')
plt.plot(Xg,Con_test,'-')
plt.axis([0.0, 3.0, -0.2, 1.2])
plt.show()
```

### Third-order upwind scheme:
For a positive advenction velocity, the derivative is calculated according to:

$$u \frac{\partial C}{\partial x} = u_{i} \frac{2C_{i+1} + 3C_{i} - 6C_{i-1} + C_{i-2}}{6 \Delta x}.$$

For a negative advection velocity, the derivative is calculated according to:

$$u \frac{\partial C}{\partial x} = u_{i} \frac{-C_{i+2} + 6C_{i+1} -3 C_{i} - 2 C_{i-1} }{6 \Delta x}.$$

### Part 2: 2D simulation of convection-diffusion transport. 

In the next practice, we will work on 2D simulation using third-order upwind scheme. As shown in the pre-class assginment, the 2D convection-diffusion equation is 

$$ \frac{\partial C}{\partial t} = D \bigg( \frac{\partial^2 C}{\partial x^2} + \frac{\partial^2 C}{\partial y^2} \bigg) - \bigg( u \frac{\partial C}{\partial x} + v \frac{\partial C}{\partial y} \bigg). $$

### In the code below, the swirl of velocity field is provided. The initial concentration is also provided. Complete the code below and do the simulation.

```python
############################################################
# Code 3: 2D upwind scheme
############################################################


# load libraries
import math
import numpy as np
import time 

import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

from IPython.display import display, clear_output

# parameter
tm = 0.0
Dfs = 2.0e-1

# spatial discretization
nx = 101
Xg = np.linspace(0, 1, nx)

ny = 101
Yg = np.linspace(0, 1, ny)

# meshgrid generate the x and y coordinates of the 2D grid system
xmsh, ymsh = np.meshgrid(Xg, Yg)

dx = Xg[2] - Xg[1]
dy = Yg[2] - Yg[1]
dt = 1.0e-6
# dt = 0.2*min(dx,dy)**2/Dfs

print(dt)


x0 = 0.25
y0 = 0.25


# lists for storing results
dSq = np.zeros((nx,ny))
Con_o = np.zeros((nx,ny))
Con_n = np.zeros((nx,ny))
Lap = np.zeros((nx,ny))
Cnvc = np.zeros((nx,ny))
Cnvc_u = np.zeros((nx,ny))
Cnvc_v = np.zeros((nx,ny))

# initial condition
t0 = 0.0005
dSq[0:nx,0:ny] = (xmsh[0:nx,0:ny]-x0)**2 + (ymsh[0:nx,0:ny]-y0)**2    
Con_o[0:nx,0:ny] = math.sqrt(t0/t0)*np.exp(-dSq[0:nx,0:ny]/4/2.0/t0)   

# velocity
u1 = np.zeros((nx,ny))
v1 = np.zeros((nx,ny))
dSq[0:nx,0:ny] = np.sqrt((xmsh[0:nx,0:ny]-0.5)**2 + (ymsh[0:nx,0:ny]-0.5)**2)*500

for i in range(nx):
    for j in range(ny):
        
        Xpj = xmsh[i,j]-0.5
        Ypj = ymsh[i,j]-0.5
        if abs(Xpj) == 0:
            Xpj = 1.0e-7
            
        angle = math.atan2(Ypj, Xpj)
        
        u1[i,j] = +dSq[i,j]*math.sin(angle)
        v1[i,j] = -dSq[i,j]*math.cos(angle)
        

print(np.shape(u1))
print(np.shape(xmsh))

# time stepping
for iter in range(8000):
    
    # calculat the advection term at each grid point
    for n in range(2 ,nx-2 ):         # <== fill the blank
        for m in range(2 ,ny-2 ):     # <== fill the blank
            
            # take derivative along the n-index 
            if v1[n,m] >= 0.0:
                Cnvc_v[n,m] = v1[n,m]*(2*Con_o[n+1,m]+3*Con_o[n,m]-6*Con_o[n-1,m]+Con_o[n-2,m])/(6*dx)    # <== fill the blank
            else:
                Cnvc_v[n,m] = v1[n,m]*(-Con_o[n+2,m]+6*Con_o[n+1,m]-3*Con_o[n,m]-2*Con_o[n-1,m])/(6*dx)      # <== fill the blank
            
            # take derivative along the m-index
            if u1[n,m] >= 0.0:
                Cnvc_u[n,m] = u1[n,m]*(2*Con_o[n,m+1]+3*Con_o[n,m]-6*Con_o[n,m-1]+Con_o[n,m-2])/(6*dx)    # <== fill the blank
            else:
                Cnvc_u[n,m] = u1[n,m]*(-Con_o[n,m+2]+6*Con_o[n,m+1]-3*Con_o[n,m]-2*Con_o[n,m-1])/(6*dx)     # <== fill the blank

    Cnvc[2:nx-2,2:ny-2 ] = Cnvc_u[2:nx-2 ,2:ny-2 ] + Cnvc_v[2:nx-2,2:ny-2 ]  # <== fill the blank
    
    

    # diffusion part (Laplace operator) 
    Lap[2:nx-2 ,2:ny-2 ] = (Con_o[1:nx-3,2:ny-2 ] - 2*Con_o[2:nx-2 ,2:ny-2 ] + Con_o[3:nx-1 ,2:ny-2 ])/dx**2 + (Con_o[2:nx-2,1:ny-3 ] - 2*Con_o[2:nx-2 ,2:ny-2 ] + Con_o[2:nx-2 ,3:ny-1 ])/dy**2         # <== fill the blank


    
    # update the concentration using vectorized calculation
    Con_n[2:nx-2,2:nx-2 ] = Con_o[2:nx-2,2:nx-2] + dt*Dfs*Lap[2:nx-2,2:nx-2 ] - dt*Cnvc[2:nx-2,2:ny-2]          # <== fill the blank
    
    
    # no-flux boundary condition
    Con_n[2:nx-2,1] = Con_n[2:nx-2,2]        # <== fill the blank
    Con_n[2:nx-2,0] = Con_n[2:nx-2,2]        # <== fill the blank
    
    Con_n[2:nx-2,ny-2] = Con_n[2:nx-2,ny-3]     # <== fill the blank
    Con_n[2:nx-2,ny-1] = Con_n[2:nx-2,ny-3]     # <== fill the blank  
    
    Con_n[0:nx-1,1] = Con_n[0:nx-1,2]        # <== fill the blank
    Con_n[0:nx-1,0] = Con_n[0:nx-1,2 ]        # <== fill the blank
    
    Con_n[0:nx-1,ny-2] = Con_n[0:nx-1, ny-3]     # <== fill the blank
    Con_n[0:nx-1,ny-1] = Con_n[0:nx-1, ny-3]     # <== fill the blank    
    
    # replace C0 with C 
    Con_o[0:nx,0:ny] = Con_n[0:nx,0:ny]  
    
    # update simulation time
    tm = tm + dt


    if iter%40 == 0:
        clear_output(wait=True) # Clear output for dynamic display
        
        fig = plt.figure(figsize=(5, 5), dpi=100)
        plt.contourf(xmsh,ymsh,Con_o, 40, cmap=cm.jet)
        plt.clim(-0.05,1.05)
        plt.colorbar()
        

        plt.quiver(xmsh[::5, ::5], ymsh[::5, ::5], u1[::5, ::5], v1[::5, ::5], color='w')
        plt.axes().set_aspect('equal')
        plt.show()
        time.sleep(0.02)         # Sleep for half a second to slow down the animation
        
```
